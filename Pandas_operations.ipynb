## **Make changes to the data**

```python
df.drop('Total',inplace=True,axis=1)
df.head(3)
```

```python

# way1
#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
#df.head()

#df=df.drop(columns=['Total'])
#df.head()

#way2
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

#rearranging the columns
cols = list(df.columns)
df = df[cols[0:4]+[cols[-1]] +cols[4:12]]

df.head(5)
```

## **Saving our data**

```python
df.to_csv("modified_pokemon.csv",index=False)
df.to_excel("modified_pokemon.xlsx",index=False)
df.to_csv("modified_pokemon.txt",sep='\t')
```

## **Filtering Data**

```python
df.loc[(df["Type 1"]=="Grass") | (df["Type 2"]=="Poison")]  # or 

new_df=df.loc[(df["Type 1"]=="Grass") & (df["Type 2"]=="Poison") & (df["HP"]>70)] # and

new_df.reset_index(drop=True,inplace=True)
new_df
```

```python
df.loc[~df['Name'].str.contains("Mega ")]   # all names which don't contains 'mega'
```

```python
import re
df.loc[df['Type 1'].str.contains("Fire|Grass",regex=True)] 
#ignore case 
df.loc[df['Type 1'].str.contains("fire|grass",flags=re.I,regex=True)] 
```

```python
df.loc[df['Name'].str.contains("^pi[a-z]*",flags=re.I,regex=True)] 
```

## **Conditional changes**

```python
df.loc[df['Type 1']=='Fire','Type 1']='Flame'

df.loc[df['Type 1']=='Flame','Legendary']=True

# multiple change
df.loc[df['Total']>500,['Generation','Legendary']]=['Test1','Test2']
df
```

## **Aggregate statistics (Groupby)**

```python
df=pd.read_csv("modified_pokemon.csv")

df.groupby('Type 1').mean()

df.groupby('Type 1').sum()

df.groupby('Type 1').count()

```

```python
# which type has highest attack mean value
df.groupby('Type 1').mean().sort_values("Attack",ascending=False)
```
